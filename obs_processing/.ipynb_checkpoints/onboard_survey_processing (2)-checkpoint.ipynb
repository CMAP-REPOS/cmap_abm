```python
def add_percentage_to_crosstab(ct):
    total_counts = ct['All'][:-1].sum()
    ct['Percent'] = ct.apply(lambda row: round(100 * row['All'] / total_counts, 2), axis=1)
    ct.loc['Percent'] = ct.apply(lambda col: round(100 * col['All'] / total_counts, 2), axis=0)
    ct.loc['Percent', 'Percent'] = pd.NA
    return ct
```

```python
def make_percentage_bar_plot(variable, title, xlabel, xlabel_rotation):
    grouped_df = obs_df.groupby(variable)["NewOriginating Rides"].agg(sum)
    total_weight = grouped_df.sum()
    grouped_df["percent"] = grouped_df/total_weight*100
    plt.figure(figsize=(10,7))
    grouped_df["percent"].plot.bar()
    plt.xlabel(xlabel, fontsize=16)
    plt.xticks(rotation=xlabel_rotation)
    plt.ylabel("Weighted Percent", fontsize=16)
    plt.title(title, fontsize=16)
    plt.show()
```

```python
# Code surveyed routes as premium or local

def code_service_type(route):
    if(route in [range(1, 88 + 1), range(92, 99 + 1), range(152, 156), 291, 1001 + 1]):
        return 1
    if(route in [90, 100, 190, 200, 290, 203, range(193, 195 + 1)]):
        return 2
    if(route == 0):
        return np.NaN
    return 1

obs_df['ROUTE_SURVEYED_service'] = obs_df['Route'].apply(lambda route: code_service_type(route))
```

```python
def code_purpose(value):
    if (value == 1):
        return 'Work'
    elif (value == 2):
        return 'Other work'
    elif (value == 3):
        return 'Home'
    elif (value == 4):
        return 'University'
    elif (value == 5):
        return 'Airport'
    elif (value == 6):
        return 'Recreation'
    elif (value == 7):
        return 'Medical'
    elif (value == 8):
        return 'Social'
    elif (value == 9):
        return 'Personal business'
    elif (value == 10):
        return 'Escort'
    elif (value == 11):
        return 'Shopping'
    elif (value == 12):
        return 'Dine out'
    elif (value == 13):
        return 'K-12 school'
    elif (value == 14):
        return 'Hotel'
    elif (value == 15):
        return 'Sporting event'
    elif (value == 16):
        return 'Other'
```

```python
obs_df['ORIG_PURPOSE'] = obs_df['Q5'].apply(lambda value: code_purpose(value))

obs_df['DEST_PURPOSE'] = obs_df['Q6'].apply(lambda value: code_purpose(value))
```

```python
def determine_tour_purpose(orig_purpose, dest_purpose):
    if (orig_purpose == 'Work') & (dest_purpose == 'Work'):
        return 'Work sub-tour'
    if (orig_purpose == 'Work') | (dest_purpose == 'Work'):
        return 'Work'
    if (orig_purpose == 'Other work') | (dest_purpose == 'Other work'):
        return 'Maintenance' # edit to other maintenance
    if (orig_purpose == 'University') | (dest_purpose == 'University'):
        return 'University'
    if (orig_purpose == 'K-12 school') | (dest_purpose == 'K-12 school'):
        return 'School'
    if (orig_purpose == 'Escort') | (dest_purpose == 'Escort'):
        return 'Maintenance'
    if (orig_purpose == 'Shopping') | (dest_purpose == 'Shopping'):
        return 'Maintenance'
    if (orig_purpose in ['Personal business', 'Medical']) | (dest_purpose in ['Personal business', 'Medical']):
        return 'Maintenance'
    if (orig_purpose == 'Social') | (dest_purpose == 'Social'):
        return 'Discretionary'
    if (orig_purpose in ['Recreation', 'Sporting event', 'Other']) | (dest_purpose in ['Recreation', 'Sporting event', 'Other']):
        return 'Discretionary'
    if (orig_purpose == 'Dine out') | (dest_purpose == 'Dine out'):
        return 'Discretionary'
    if (orig_purpose == 'Home') & (dest_purpose != 'Home'):
        return dest_purpose
    if (orig_purpose != 'Home') & (dest_purpose == 'Home'):
        return orig_purpose
    if (orig_purpose == 'Home') & (dest_purpose == 'Home'):
        return 'Discretionary'
    if orig_purpose == dest_purpose:
        return orig_purpose
    return dest_purpose

obs_df['TOUR_PURPOSE'] = obs_df.apply(lambda row: determine_tour_purpose(row['ORIG_PURPOSE'], row['DEST_PURPOSE']), axis=1)

# home-home as other discrectionary
# Social/visiting vs. other discretionary
# Recreation & sporting in "other discretionary"
```

```python
# Derive observed/surveyed hour (as a sub for departure hour)

# pd.to_datetime(obs_df['StartTime']).head()
obs_df['surveyed_hour'] = pd.to_datetime(obs_df['StartTime']).dt.hour

obs_df['surveyed_hour'].value_counts(dropna=False).sort_index(level='surveyed_hour')
```

```python
# Impute work tour purpose
# Limit to workers, tour purpose is maintenance/discretionary
# limit to weekdays? need to ask -- Day_Type_2 == 1

obs_df['TOUR_PURPOSE_INIT'] = obs_df['TOUR_PURPOSE']
obs_df['imputed_tour_purpose'] = ''

# calculate utility work tour based on model coefficients
obs_df['uWTP'] = -.8340969

# original tour purpose
obs_df.loc[(obs_df['TOUR_PURPOSE_INIT'] == 'Discretionary'), 'uWTP'] = obs_df['uWTP'] + (1.61246)


# surveyed hour
obs_df.loc[(obs_df['surveyed_hour'].isin(range(9, 15 + 1))), 'uWTP'] = obs_df['uWTP'] + (-2.680612)
obs_df.loc[(obs_df['surveyed_hour'] == 16), 'uWTP'] = obs_df['uWTP'] + (-1.775932)
obs_df.loc[(obs_df['surveyed_hour'] == 17), 'uWTP'] = obs_df['uWTP'] + (-1.270239)
obs_df.loc[(obs_df['surveyed_hour'] == 18), 'uWTP'] = obs_df['uWTP'] + (0) # not significant
obs_df.loc[(obs_df['surveyed_hour'] == 19), 'uWTP'] = obs_df['uWTP'] + (0) # not significant
obs_df.loc[(obs_df['surveyed_hour'] == 20), 'uWTP'] = obs_df['uWTP'] + (-1.343958)
obs_df.loc[(obs_df['surveyed_hour'] == 21), 'uWTP'] = obs_df['uWTP'] + (0) # not significant

# access mode
obs_df.loc[(obs_df['Q31'] > 0) & (obs_df['TOUR_MODE'] == 'PNR'), 'uWTP'] = obs_df['uWTP'] + (0) # not significant
obs_df.loc[(obs_df['Q31'] > 0) & (obs_df['TOUR_MODE'] == 'KNR'), 'uWTP'] = obs_df['uWTP'] + (0) # not significant


# calculate probability
obs_df['expuWTP'] = np.exp(obs_df['uWTP'])
obs_df['probWTP'] = obs_df['expuWTP']/(1 + obs_df['expuWTP'])

print(obs_df['probWTP'].value_counts(dropna=False))

# Generate random number
np.random.seed(30)
obs_df['randn'] = np.random.uniform(0, 1, obs_df.shape[0])

obs_df.loc[(obs_df['randn'] < obs_df['probWTP']), 'imputed_tour_purpose'] = 'Work'
obs_df.loc[(obs_df['randn'] >= obs_df['probWTP']), 'imputed_tour_purpose'] = 'replace'


print(np.mean(obs_df['probWTP']))
print(obs_df[obs_df['imputed_tour_purpose'] == 'Work'].shape[0]/
      (obs_df[obs_df['imputed_tour_purpose'] == 'Work'].shape[0] + 
       obs_df[obs_df['imputed_tour_purpose'] == 'replace'].shape[0]))

obs_df.loc[(obs_df['Q31'] > 0) & 
           (obs_df['imputed_tour_purpose'] == 'Work') & 
           (obs_df['TOUR_PURPOSE_INIT'].isin(['Maintenance', 'Discretionary'])), 'TOUR_PURPOSE'] = obs_df['imputed_tour_purpose']

obs_df.groupby(['TOUR_PURPOSE'])['NewOriginating Rides'].agg(sum).sort_index(level='TOUR_PURPOSE')

```

```python
# Auto sufficiency in model is cars/adults -- survey only gives hh size and num workers

# calculate utility greater/equal adults based on model coefficients
obs_df['uGEA'] = 0 # constant not significant

# num autos
obs_df.loc[(obs_df['num_autos'] == 2), 'uGEA'] = obs_df['uGEA'] + (-3.211687)
obs_df.loc[(obs_df['num_autos'] >= 3), 'uGEA'] = obs_df['uGEA'] + (-3.820387)

# num workers
obs_df.loc[(obs_df['Q31'] == 1), 'uGEA'] = obs_df['uGEA'] + (0) # not significant
obs_df.loc[(obs_df['Q31'] == 2), 'uGEA'] = obs_df['uGEA'] + (1.542208)
obs_df.loc[(obs_df['Q31'] >= 3), 'uGEA'] = obs_df['uGEA'] + ( 3.68089)

# hh size
obs_df.loc[(obs_df['Q30'] == 3), 'uGEA'] = obs_df['uGEA'] + (-.8128553)
obs_df.loc[(obs_df['Q30'] == 4), 'uGEA'] = obs_df['uGEA'] + (-.6252058)
obs_df.loc[(obs_df['Q30'] >= 5), 'uGEA'] = obs_df['uGEA'] + (-.781304)

# income
obs_df.loc[(obs_df['Q38'].isin([2, 3])), 'uGEA'] = obs_df['uGEA'] + (-1.765795)
obs_df.loc[(obs_df['Q38'].isin([4, 5])), 'uGEA'] = obs_df['uGEA'] + (0) # not significant
obs_df.loc[(obs_df['Q38'].isin(range(6, 10 + 1))), 'uGEA'] = obs_df['uGEA'] + (0) # not significant
obs_df.loc[(obs_df['Q38'].isin(range(11, 13 + 1))), 'uGEA'] = obs_df['uGEA'] + (0) # not significant
obs_df.loc[(obs_df['Q38'] == 14), 'uGEA'] = obs_df['uGEA'] + (-1.530027)

# drive access
obs_df.loc[(obs_df['Q8'] == 4), 'uGEA'] = obs_df['uGEA'] + (-.7000864)

# calculate probability
obs_df['expuGEA'] = np.exp(obs_df['uGEA'])
obs_df['probAS2'] = obs_df['expuGEA']/(1 + obs_df['expuGEA'])
# obs_df['uGEA'].value_counts(dropna=False)
# print(obs_df['probAS2'].value_counts(dropna=False))

# Generate random number
np.random.seed(30)
obs_df['randn'] = np.random.uniform(0, 1, obs_df.shape[0])

obs_df.loc[(obs_df['randn'] < obs_df['probAS2']), 'imputed_auto_suff'] = 2
obs_df.loc[(obs_df['randn'] >= obs_df['probAS2']), 'imputed_auto_suff'] = 1


obs_df.loc[(obs_df['AUTO_SUFF'] == 1), 'AUTO_SUFF'] = obs_df['imputed_auto_suff']

print(np.mean(obs_df['probAS2']))
print(obs_df[obs_df['imputed_auto_suff'] == 2].shape[0]/
      (obs_df[obs_df['imputed_auto_suff'] == 2].shape[0] + 
       obs_df[obs_df['imputed_auto_suff'] == 1].shape[0]))
print(obs_df['imputed_auto_suff'].value_counts(dropna=False))
```
