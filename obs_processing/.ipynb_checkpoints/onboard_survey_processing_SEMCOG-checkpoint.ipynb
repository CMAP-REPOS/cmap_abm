```python
def add_percentage_to_crosstab(ct):
    total_counts = ct['All'][:-1].sum()
    ct['Percent'] = ct.apply(lambda row: round(100 * row['All'] / total_counts, 2), axis=1)
    ct.loc['Percent'] = ct.apply(lambda col: round(100 * col['All'] / total_counts, 2), axis=0)
    ct.loc['Percent', 'Percent'] = pd.NA
    return ct
```

```python
def read_matrix_from_omx_file(file, mode):
    matrix_df = pd.DataFrame(file[mode][:])
    if ("ZoneID" in file.list_mappings()):
        inv_row_map = file.mapping("ZoneID")
        row_map = {taz1: taz2 for taz2, taz1 in inv_row_map.items()}
        inv_col_map = file.mapping("ZoneID")
        col_map = {taz1: taz2 for taz2, taz1 in inv_col_map.items()}
        matrix_df.rename(index=row_map, columns=col_map, inplace=True)
    else:
        row_col_map = {idx: idx+1 for idx in matrix_df.columns}
        matrix_df.rename(index=row_col_map, columns=row_col_map, inplace=True)
        
    # Exporting null values from a TransCAD skim can give either really large or really small values... setting these to 0
    matrix_df = matrix_df.applymap(lambda x: 0 if (x < .00001 or x > 100000) else x)
    return matrix_df
```

```python
def determine_tour_purpose(orig_purpose, dest_purpose):
    if (orig_purpose == 'Work') & (dest_purpose == 'Work'):
        return 'At Work'
    if (orig_purpose == 'Work') | (dest_purpose == 'Work'):
        return 'Work'
    if (orig_purpose == 'University') | (dest_purpose == 'University'):
        return 'University'
    if (orig_purpose == 'School') | (dest_purpose == 'School'):
        return 'School'
    if (orig_purpose == 'Escorting') | (dest_purpose == 'Escorting'):
        return 'Escorting'
    if (orig_purpose == 'Shopping') | (dest_purpose == 'Shopping'):
        return 'Shopping'
    if (orig_purpose == 'Other Maintenance') | (dest_purpose == 'Other Maintenance'):
        return 'Other Maintenance'
    if (orig_purpose == 'Social/Rec') | (dest_purpose == 'Social/Rec'):
        return 'Social/Rec'
    if (orig_purpose == 'Eat Out') | (dest_purpose == 'Eat Out'):
        return 'Eat Out'
    if (orig_purpose == 'Home') & (dest_purpose != 'Home'):
        return dest_purpose
    if (orig_purpose != 'Home') & (dest_purpose == 'Home'):
        return orig_purpose
    if orig_purpose == dest_purpose:
        return orig_purpose
    if (orig_purpose == 'Home') & (dest_purpose == 'Home'):
        return 'Home'
    return dest_purpose

obs_df['TOUR_PURPOSE'] = obs_df.apply(lambda row: determine_tour_purpose(row['ORIG_PURPOSE'], row['DEST_PURPOSE']), axis=1)
```

```python
def determine_transit_mode(route, route_code):
    mode = pd.NA
    if 'DPM' in route_code:
        mode = "Detroit People Mover"
    if 'DDT' in route_code:
        mode = "Bus-Local"
    if 'AAA' in route_code:
        mode = "Bus-Local"
    if 'SMD' in route_code:  
        mode = "Bus-Local"
    if 'UMT' in route_code:
        mode = 'Bus-Local'
    if 'QLN' in route_code:
        mode = 'Street Car'
    if 'BWT' in route_code:
        mode = 'Bus-Local'
    if 'LET' in route_code:
        mode = 'Bus-Local'
    if 'FLT' in route_code:
        mode = 'Bus-Local'
    if 'WDS' in route_code:
        mode = 'Bus-Local'
    if 'oth' in route_code:
        mode = 'Bus-Local'  # Assume 'Other' is bus-local
    if pd.isna(mode):
        print("got NA mode for", route, route_code)
    if mode == 'Bus-Local':
        if 'exp' in route.lower():
            return 'Bus-Express'
        else:
            return 'Bus-Local'
    return mode

# obs_df['TRANSIT_MODE'] = obs_df['ROUTE_SURVEYED.Code.'].apply(lambda route_code: determine_transit_mode(route_code))
obs_df['TRANSIT_MODE'] = obs_df.apply(lambda row: 
                                      determine_transit_mode(row['ROUTE_SURVEYED'],row['ROUTE_SURVEYED.Code.']), axis=1)
```

### Transit Access/Egress Mode
Broken down into 3 modes:
* Walk (includes all non-motorized modes like bike as well as E scooter and Scooter)
* PNR 
* KNR (includes Taxi, work shuttle)
* (TNC either included as own mode or grouped into KNR)

```python
orig_transport_df = obs_df.groupby(['ORIGIN_TRANSPORT','ORIGIN_TRANSPORT.Code.'])['Updated_LinkedWeightFactor'].agg(sum).sort_index(level='ORIGIN_TRANSPORT.Code.').to_frame()
orig_transport_df['Percent'] = orig_transport_df['Updated_LinkedWeightFactor'] / orig_transport_df['Updated_LinkedWeightFactor'].sum() * 100
orig_transport_df
```

```python
def determine_access_egress_mode(access_egress_code):
    if int(access_egress_code) in [1,2,3,4,5,99]:
        return 'Walk'
    if int(access_egress_code) in [7,8,9]:
        return 'PNR'
    if int(access_egress_code) in [6,10,11,12]:
        return 'KNR'

def determine_home_mode(row):
    if row['ORIG_PURPOSE'] == 'Home':
        return determine_access_egress_mode(row['ORIGIN_TRANSPORT.Code.'])
    if row['DEST_PURPOSE'] == 'Home':
        return determine_access_egress_mode(row['DESTIN_TRANSPORT.Code.'])
    return pd.NA
    
def determine_tour_mode(row):
    if (row['ORIG_MODE'] == 'PNR') | (row['DEST_MODE'] == 'PNR'):
        return 'PNR'
    if (row['ORIG_MODE'] == 'KNR') | (row['DEST_MODE'] == 'KNR'):
        return 'KNR'
    return 'Walk'

obs_df['ORIG_MODE'] = obs_df['ORIGIN_TRANSPORT.Code.'].apply(lambda access_code: determine_access_egress_mode(access_code))
obs_df['DEST_MODE'] = obs_df['DESTIN_TRANSPORT.Code.'].apply(lambda egress_code: determine_access_egress_mode(egress_code))
obs_df['HOME_MODE'] = obs_df.apply(lambda row: determine_home_mode(row), axis=1)
obs_df['TOUR_MODE'] = obs_df.apply(lambda row: determine_tour_mode(row), axis=1)
```

```python
def determine_access_egress_mode_with_TNC(access_egress_code):
    if int(access_egress_code) in [1,2,3,4,5,99]:
        return 'Walk'
    if int(access_egress_code) in [7,8,9]:
        return 'PNR'
    if int(access_egress_code) in [6,11]:
        return 'KNR'
    if int(access_egress_code) in [10,12]:
        return 'TNC'
    
def determine_home_mode_with_TNC(row):
    if row['ORIG_PURPOSE'] == 'Home':
        return determine_access_egress_mode_with_TNC(row['ORIGIN_TRANSPORT.Code.'])
    if row['DEST_PURPOSE'] == 'Home':
        return determine_access_egress_mode_with_TNC(row['DESTIN_TRANSPORT.Code.'])
    return pd.NA

def determine_tour_mode_with_TNC(row):
    if (row['ORIG_MODE_INCL_TNC'] == 'TNC') | (row['DEST_MODE_INCL_TNC'] == 'TNC'):
        return 'TNC'
    if (row['ORIG_MODE_INCL_TNC'] == 'PNR') | (row['DEST_MODE_INCL_TNC'] == 'PNR'):
        return 'PNR'
    if (row['ORIG_MODE_INCL_TNC'] == 'KNR') | (row['DEST_MODE_INCL_TNC'] == 'KNR'):
        return 'KNR'
    return 'Walk'

obs_df['ORIG_MODE_INCL_TNC'] = obs_df['ORIGIN_TRANSPORT.Code.'].apply(lambda access_code: determine_access_egress_mode_with_TNC(access_code))
obs_df['DEST_MODE_INCL_TNC'] = obs_df['DESTIN_TRANSPORT.Code.'].apply(lambda egress_code: determine_access_egress_mode_with_TNC(egress_code))
obs_df['HOME_MODE_INCL_TNC'] = obs_df.apply(lambda row: determine_home_mode_with_TNC(row), axis=1)
obs_df['TOUR_MODE_INCL_TNC'] = obs_df.apply(lambda row: determine_tour_mode_with_TNC(row), axis=1)
```

```python
geod = pyproj.Geod(ellps="WGS84")
meters_to_miles = .000621371

def determine_origin_distance(row):
    lon1 = row['ORIG_LON']
    lat1 = row['ORIG_LAT']
    lon2 = pd.NA
    lat2 = pd.NA
    if row['PREV_TRANSFERS.Code.'] == 0:
        lat2 = row['STOP_ON..LAT.']
        lon2 = row['STOP_ON..LONG.']
    if row['PREV_TRANSFERS.Code.'] == 1:
        lat2 = row['PREV_TRAN_1_ON_BUS..LAT.']
        lon2 = row['PREV_TRAN_1_ON_BUS..LONG.']
    if row['PREV_TRANSFERS.Code.'] == 2:
        lat2 = row['PREV_TRAN_2_ON_BUS..LAT.']
        lon2 = row['PREV_TRAN_2_ON_BUS..LONG.']
    if row['PREV_TRANSFERS.Code.'] == 3:
        lat2 = row['PREV_TRAN_3_ON_BUS..LAT.']
        lon2 = row['PREV_TRAN_3_ON_BUS..LONG.']
    
    if pd.isna(lon1) | pd.isna(lat1) | pd.isna(lat2) | pd.isna(lon2):
        return pd.NA
    else:
        forward_azi, back_azi, dist_in_meters = geod.inv(lon1, lat1, lon2, lat2)
        return dist_in_meters * meters_to_miles

def determine_destination_distance(row):
    lon1 = row['DEST_LON']
    lat1 = row['DEST_LAT']
    lon2 = pd.NA
    lat2 = pd.NA
    if row['NEXT_TRANSFERS.Code.'] == 0:
        lat2 = row['STOP_OFF..LAT.']
        lon2 = row['STOP_OFF..LONG.']
    if row['NEXT_TRANSFERS.Code.'] == 1:
        lat2 = row['NEXT_TRAN_1_OFF_BUS..LAT.']
        lon2 = row['NEXT_TRAN_1_OFF_BUS..LONG.']
    if row['NEXT_TRANSFERS.Code.'] == 2:
        lat2 = row['NEXT_TRAN_2_OFF_BUS..LAT.']
        lon2 = row['NEXT_TRAN_2_OFF_BUS..LONG.']
    if row['NEXT_TRANSFERS.Code.'] == 3:
        lat2 = row['NEXT_TRAN_3_OFF_BUS..LAT.']
        lon2 = row['NEXT_TRAN_3_OFF_BUS..LONG.']
    
    if pd.isna(lon1) | pd.isna(lat1) | pd.isna(lat2) | pd.isna(lon2):
        return pd.NA
    else:
        forward_azi, back_azi, dist_in_meters = geod.inv(lon1, lat1, lon2, lat2)
        return dist_in_meters * meters_to_miles
    
def calculate_orig_dest_distance(row):
    lon1 = row['ORIG_LON']
    lat1 = row['ORIG_LAT']
    lon2 = row['DEST_LON']
    lat2 = row['DEST_LAT']
    
    if pd.isna(lon1) | pd.isna(lat1) | pd.isna(lat2) | pd.isna(lon2):
        return pd.NA
    else:
        forward_azi, back_azi, dist_in_meters = geod.inv(lon1, lat1, lon2, lat2)
        return dist_in_meters * meters_to_miles
    
obs_df['ORIG_DIST'] = obs_df.apply(lambda row: determine_origin_distance(row), axis=1)
obs_df['DEST_DIST'] = obs_df.apply(lambda row: determine_destination_distance(row), axis=1)
obs_df['ORIG_TO_DEST_DIST'] = obs_df.apply(lambda row: calculate_orig_dest_distance(row), axis=1)
# obs_df[:10].apply(lambda row: determine_origin_distance(row), axis = 1)
```

```python
transfer_route_codes = [
    'TRIP_FIRST_ROUTE.Code.',
    'TRIP_SECOND_ROUTE.Code.',
    'TRIP_THIRD_ROUTE.Code.',
    'ROUTE_SURVEYED.Code.',
    'TRIP_NEXT_ROUTE.Code.',
    'TRIP_AFTER_ROUTE.Code.',
    'TRIP_3RD_ROUTE.Code.']

transfer_routes = [
    'TRIP_FIRST_ROUTE',
    'TRIP_SECOND_ROUTE',
    'TRIP_THIRD_ROUTE',
    'ROUTE_SURVEYED',
    'TRIP_NEXT_ROUTE',
    'TRIP_AFTER_ROUTE',
    'TRIP_3RD_ROUTE']

def determine_survey_set_type(row):
    used_local = False
    used_premium = False
    
    for transfer_route, transfer_code in zip(transfer_routes, transfer_route_codes):
        if (pd.isna(row[transfer_route])) | (pd.isna(row[transfer_code])):
            continue
        transfer_mode = determine_transit_mode(row[transfer_route], row[transfer_code])
#         print(transfer_mode)
        if (transfer_mode == 'Bus-Local') | (transfer_mode == 'Bus-Express'):
            used_local = True
        if (transfer_mode == 'Street Car') | (transfer_mode == 'Detroit People Mover'):
            used_premium = True
            
    if (used_local & used_premium):
        return 'Local+Premium'
    elif used_premium:
        return 'Premium Only'
    elif used_local:
        return 'Local Only'

obs_df['SURVEY_SET_TYPE'] = obs_df.apply(lambda row: determine_survey_set_type(row), axis=1)
# obs_df[:10].apply(lambda row: determine_survey_set_type(row), axis=1)
```

```python
def make_comparison_bar_plot(variable, title, xlabel, xlabel_rotation, dfs, sources, sort_by=None):
    all_dfs = []
    for df, source in zip(dfs, sources):
        if sort_by is not None:
            df_groups = df.groupby([variable, sort_by])['Updated_LinkedWeightFactor'].agg(sum)
            df_groups = df_groups.sort_index(level=sort_by)
        else:
            df_groups = df.groupby(variable)['Updated_LinkedWeightFactor'].agg(sum)
        groups_df = df_groups.to_frame()
        groups_df["percent"] = groups_df['Updated_LinkedWeightFactor'] / groups_df['Updated_LinkedWeightFactor'].sum() * 100
        groups_df['source'] = source
        all_dfs.append(groups_df)
    all_df = pd.concat(all_dfs).reset_index()

    plt.figure(figsize=(10,7))
    sns.barplot(data=all_df, x=variable, y='percent', hue='source')
    plt.xlabel(xlabel, fontsize=16)
    plt.xticks(rotation=xlabel_rotation)
    plt.ylabel("Weighted Percent", fontsize=16)
    plt.title(title, fontsize=16)
    plt.show()
```

```python
def evaluate_transit_services(dfs, descriptions):
    
    tot_linked_trips = obs_df['Updated_LinkedWeightFactor'].sum()
    print("Total linked transit trips:", int(tot_linked_trips))
    
    for df, desc in zip(dfs, descriptions):
        tot_seg_linked_trips = df['Updated_LinkedWeightFactor'].sum()
        print(desc, "linked transit trips:", int(tot_seg_linked_trips))
        portion = tot_seg_linked_trips / tot_linked_trips
        print(desc, "trips account for ", round(portion*100,2), '% of total transit ridership')
    
    make_comparison_bar_plot(
        variable='TOUR_PURPOSE',
        title="Tour Purpose",
        xlabel="Tour Purpose",
        xlabel_rotation=45,
        dfs=dfs,
        sources=descriptions)

    make_comparison_bar_plot(
        variable='STUDENT_STATUS',
        title="Student Status",
        xlabel="Student Status",
        xlabel_rotation=90,
        dfs=dfs,
        sources=descriptions)

    make_comparison_bar_plot(
        variable='EMPLOYMENT_STATUS',
        title="Employment Status",
        xlabel="Employment Status",
        xlabel_rotation=90,
        dfs=dfs,
        sources=descriptions)
    
    make_comparison_bar_plot(
        variable='AGE',
        title="Age Distribution",
        xlabel="Age Group, code",
        xlabel_rotation=0,
        dfs=dfs,
        sources=descriptions,
        sort_by='AGE.Code.')
    
    make_comparison_bar_plot(
        variable='INCOME',
        title="Income Distribution",
        xlabel="Income Group, code",
        xlabel_rotation=90,
        dfs=dfs,
        sources=descriptions,
        sort_by='INCOME.Code.')
    
    make_comparison_bar_plot(
        variable='AUTO_SUFF',
        title="Auto Sufficiency",
        xlabel="Auto Sufficiency",
        xlabel_rotation=90,
        dfs=dfs,
        sources=descriptions)

    # alighting distance
    plt.figure(figsize=(10,7))
    for df, desc in zip(dfs, descriptions):
        df['ORIG_TO_DEST_SKIM_DIST'].hist(
        bins=np.linspace(0,30,60),
        weights=df.loc[pd.notna(df['ORIG_TO_DEST_SKIM_DIST']),'Updated_LinkedWeightFactor'],
        alpha=0.5, density=True, range=[0,4],
        label=desc)
    plt.title("Skim Distance Origin to Destination", fontsize=16)
    plt.xlabel("Distance [Miles]", fontsize=16)
    plt.legend(loc='best')
    plt.ylabel("Normalized Counts", fontsize=16)
    plt.show()

    # boarding distances
    plt.figure(figsize=(20,7))
    plt.subplot(121)
    for df, desc in zip(dfs, descriptions):
        df['ORIG_DIST'].hist(
            bins=np.linspace(0,2,40),
            weights=df.loc[pd.notna(df['ORIG_DIST']),'Updated_LinkedWeightFactor'],
            alpha=0.5, density=True, range=[0,4],
            label=desc)
    plt.title("Distance From Origin to First Boarding Location", fontsize=16)
    plt.xlabel("Distance [Miles]", fontsize=16)
    plt.legend(loc='best')
    plt.ylabel("Normalized Counts", fontsize=16)

    plt.subplot(122)
    for df, desc in zip(dfs, descriptions):
        df['DEST_DIST'].hist(
            bins=np.linspace(0,2,40),
            weights=df.loc[pd.notna(df['ORIG_DIST']),'Updated_LinkedWeightFactor'],
            alpha=0.5, density=True, range=[0,4],
            label=desc)
    plt.title("Distance From Last Alighting to Destination", fontsize=16)
    plt.xlabel("Distance [Miles]", fontsize=16)
    plt.legend(loc='best')
    plt.ylabel("Normalized Counts", fontsize=16)
    plt.show()
    
    make_comparison_bar_plot(
        variable='TOUR_MODE_INCL_TNC',
        title="Tour Mode",
        xlabel="Tour Mode",
        xlabel_rotation=0,
        dfs=dfs,
        sources=descriptions)

    make_comparison_bar_plot(
        variable='TOTAL_TRANSFERS',
        title="Total Transfers",
        xlabel="Total Transfers",
        xlabel_rotation=0,
        dfs=dfs,
        sources=descriptions)

```
